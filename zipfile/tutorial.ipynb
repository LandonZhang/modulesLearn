```python
import zipfile
from pathlib import Path

workspace = Path()

test_file = workspace / "test.txt"
# 创建文件
test_file.touch()

with open(test_file.absolute(), "w") as f:
    f.write("Hello! Nice to meet u~\nHope u have a good day!")

print("Complete")
```

```python
game = workspace / "game.txt"
game.touch()

with open(game.absolute(), "w") as f:
    f.writelines(["Wayward is a yellow dog\n", "SHU is a good school may be..."])

print("complete")
```

```python
# 压缩文件
import zipfile

zip_path = workspace / "file_1.zip"

# 打开一个文件流
zip_file = zipfile.ZipFile(zip_path.absolute(), "w")

# 存入文件
# arcname 参数控制压缩包内的路径结构 (因为 zipfile 无法压缩文件夹,只能后天营造出这种层级感)
zip_file.write(game.absolute(), "game1.txt")
zip_file.write(test_file.absolute(), "test1.txt")
zip_file.close()
```

```python
# 添加文件进入压缩包
zip_file = zipfile.ZipFile(zip_path.absolute(), "a")

# 设置多层级结构
zip_file.write(game.absolute(), "dir1/game.txt")
zip_file.close()
```

```python
zip_path = workspace / "file_2.zip"

# 打开一个文件流(设置压缩格式)
zip_file = zipfile.ZipFile(zip_path.absolute(), "w", compression=zipfile.ZIP_DEFLATED)

# 最佳实践(上下文管理器)
with zip_file as zf:
    zf.write(game.absolute(), "game1.txt")
    zf.write(test_file.absolute(), "test1.txt")
```

```python
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "all"
```

```python
# 查看压缩文件的信息
zip_file.filename

zip_file.filelist

zip_file.namelist()

zip_file.infolist()
```

```python
# 解压全部文件
zip_path = workspace / "file_2.zip"

zip_file = zipfile.ZipFile(zip_path.absolute(), "r")

with zip_file as zf:
    zf.extractall(workspace / "extract_dir")
```

```python
# 解压特定文件

e2 = workspace / "e2"
e2.mkdir(exist_ok=True)

# 重新打开关闭的压缩文件
zip_file = zipfile.ZipFile(zip_path.absolute(), "r")

with zip_file as zf:
    namelist = zf.namelist()
    member = namelist[0]
    zf.extract(member, e2)
```

```python
# 对整个文件夹进行压缩 (zipfile 不支持)
import shutil

shutil.make_archive("e2", "zip", e2)
```

```python
# 解压缩
shutil.unpack_archive(workspace / "e2.zip", "e2_unpack")
```
