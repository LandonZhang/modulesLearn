```python
languages = ["Python", "C++", "C#", "Java"]
people = [100, 25, 21, 32]
explodes = [0.3, 0, 0, 0]

plt.pie(
    people,
    labels=languages,
    labeldistance=0.7,
    explode=explodes,
    autopct="%.2f%%",
    pctdistance=0.6,
)
```

```python
# 
a = np.random.randint(90, 100, 5)
b = np.random.randint(90, 100, 5)
c = np.random.randint(90, 100, 5)

# 
plt.plot(a.flatten(), label="line a")
plt.plot(b.flatten(), label="line b")
plt.plot(c.flatten(), label="line c")

# 
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Many lines", fontsize=12)

# 
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

x = np.arange(100)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("sine wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("cos wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("random function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("log function")

fig.set_dpi(150)
fig.suptitle("Four plots", fontsize=15)
fig.tight_layout()
```

```python
from pathlib import Path

current_path = Path().cwd()
fig_path = current_path / "pic"

fig_path.mkdir(exist_ok=True)

num = "0"
# transparent 
fig.savefig(fig_path / num, dpi=300, transparent=False)
```
