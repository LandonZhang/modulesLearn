### groupby and agg


```python
import pandas as pd

data = {
    "Name": ["zzzongzii", "Landon", "Ellie", "Joel"],
    "Email": ["1", "2", "3", "4"],
    "Hobby": ["Code", "Code", "Space", "Song"],
    "age": [21, 21, 14, 57],
}

df = pd.DataFrame(data)

groups = df.groupby(by="Hobby")

# 获得一个 group
groups.get_group("Code")
# 获得一群 group
for group_name, group_df in groups:
    print(f"Group name is {group_name}")
    print(group_df)
```

### 数据透视表

常用聚合函数

- 'sum': 求和
- 'mean': 平均值
- 'count': 计数
- 'min': 最小值
- 'max': 最大值
- 'std': 标准差
- 'var': 方差
- np.median: 中位数
- 自定义函数


```python
import pandas as pd
import numpy as np

# 创建示例数据
data = {
    "Date": [
        "2023-01-01",
        "2023-01-01",
        "2023-01-02",
        "2023-01-02",
        "2023-01-01",
        "2023-01-02",
    ],
    "City": ["北京", "上海", "北京", "上海", "广州", "广州"],
    "Product": ["A", "A", "B", "B", "A", "A"],
    "Sales": [100, 150, 200, 120, 80, 90],
    "Quantity": [10, 15, 20, 12, 8, 9],
}

df = pd.DataFrame(data)
print("原始数据:")
print(df)

# 1. 基本透视表 - 按城市和产品汇总销售额
pivot1 = pd.pivot_table(
    df,
    values="Sales",  # 汇总销售额
    index="City",  # 行索引为城市
    columns="Product",  # 列索引为产品
    aggfunc="sum",
)  # 使用求和聚合
print("\n按城市和产品汇总销售额:")
print(pivot1)
```

```python
# 2. 多个聚合函数
pivot2 = pd.pivot_table(
    df,
    values="Sales",
    index="City",
    columns="Product",
    aggfunc=["sum", "mean", "count"],
)
print("\n多个聚合函数:")
print(pivot2)
```

```python
# 3. 多个数值列
pivot3 = pd.pivot_table(
    df, values=["Sales", "Quantity"], index="City", columns="Product", aggfunc="sum"
)
print("\n多个数值列:")
print(pivot3)
```

```python
# 多个值与多个聚合函数
pivot3_1 = pd.pivot_table(
    df,
    values=["Sales", "Quantity"],
    index="City",
    columns="Product",
    aggfunc=["sum", "mean", "count"],
)
print("\n多个数值列与多个聚合函数:")
print(pivot3_1)
```

```python
# 4. 添加总计行和列
pivot4 = pd.pivot_table(
    df,
    values="Sales",
    index="City",
    columns="Product",
    aggfunc="sum",
    margins=True,  # 添加总计
    margins_name="总计",
)  # 总计行/列的名称
print("\n添加总计:")
print(pivot4)

# 5. 填充缺失值
pivot5 = pd.pivot_table(
    df, values="Sales", index="City", columns="Product", aggfunc="sum", fill_value=0
)  # 用0填充缺失值
print("\n填充缺失值:")
print(pivot5)

# 6. 多层索引
pivot6 = pd.pivot_table(
    df,
    values="Sales",
    index=[
        "Date",
        "City",
    ],  # 多层行索引(按照顺序进行聚合,先按照时间聚合,再按照城市聚合)
    columns="Product",
    aggfunc="sum",
)
print("\n多层索引:")
print(pivot6)
```

### `melt` 转化数据显示格式


```python
import pandas as pd


# 1. 基本示例 - 宽格式转长格式
print("=== 基本melt示例 ===")
wide_data = pd.DataFrame(
    {
        "Name": ["张三", "李四", "王五"],
        "语文": [85, 92, 78],
        "数学": [90, 88, 95],
        "英语": [78, 85, 82],
    }
)

print("原始宽格式数据:")
print(wide_data)

# 使用melt转换为长格式
long_data = pd.melt(
    wide_data,
    id_vars=["Name"],  # 保持Name列不变
    value_vars=["语文", "数学", "英语"],  # 要融化的列
    var_name="科目",  # 变量列名称
    value_name="成绩",
)  # 值列名称

print("\n转换后的长格式数据:")
print(long_data)
```

```python
# 2. 不指定value_vars（默认融化所有非id_vars列）
print("\n=== 不指定value_vars ===")
long_data2 = pd.melt(wide_data, id_vars=["Name"], var_name="科目", value_name="成绩")
print(long_data2)
```

```python
# 3. 多个标识符列
print("\n=== 多个标识符列 ===")
sales_data = pd.DataFrame(
    {
        "Region": ["北京", "上海", "广州"],
        "Salesperson": ["小王", "小李", "小张"],
        "Q1": [100, 150, 120],
        "Q2": [110, 140, 130],
        "Q3": [120, 160, 125],
        "Q4": [130, 170, 135],
    }
)

print("原始销售数据:")
print(sales_data)

melted_sales = pd.melt(
    sales_data,
    id_vars=["Region", "Salesperson"],  # 多个标识符
    var_name="Quarter",
    value_name="Sales",
)

print("\n融化后的销售数据:")
print(melted_sales)

# 4. 部分列融化
print("\n=== 部分列融化 ===")
partial_melt = pd.melt(
    sales_data,
    id_vars=["Region", "Salesperson"],
    value_vars=["Q1", "Q2"],  # 只融化Q1和Q2
    var_name="Quarter",
    value_name="Sales",
)

print("只融化Q1和Q2:")
print(partial_melt)  # Q3 Q4 的数据就不会再显示了
```
