```python
import torch

# 设置随机数种子以减少随机性
RANDOM_SEED = 32

# 随机数种子设置之后全局生效(只对CPU而言)
torch.manual_seed(RANDOM_SEED)
a = torch.rand(2, 3)
b = torch.rand(2, 3)

# 创建 GPU 种子
torch.cuda.manual_seed(RANDOM_SEED)
c = torch.rand(2, 3)
d = torch.rand(2, 3)

print(a, b, sep="\n**********\n")
print("\n**********\n")
print(c, d, sep="\n**********\n")
```
